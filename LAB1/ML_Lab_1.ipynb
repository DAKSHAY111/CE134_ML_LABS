```python
import nltk
from nltk.corpus import twitter_samples, stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
import matplotlib.pyplot as plt
import random
import re
import string
import numpy as np
import pandas as pd
from google.colab import drive


nltk.download('twitter_samples')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')


positiveTweets = twitter_samples.strings('positive_tweets.json')
negativeTweets = twitter_samples.strings('negative_tweets.json')

print('\nNo. of positive tweets, ', len(positiveTweets))

figure = plt.figure(figsize=(5,5))
labels = ['C++', 'Java', 'C#', 'python']
sizes = [50, 30, 6, 14]

plt.pie(sizes, labels=labels, startangle=90)
plt.axis('equal')
plt.show()

print('\n\033[92m', positiveTweets[random.randint(0,5000)])
print('\033[91m', negativeTweets[random.randint(0,5000)])

# List of common colors
# Red = ‘\033[91m’ 
# Green = ‘\033[92m’ 
# Blue = ‘\033[94m’
# Cyan = ‘\033[96m’
# White = ‘\033[97m’
# Yellow = ‘\033[93m’
# Magenta = ‘\033[95m’
# Grey = ‘\033[90m’
# Black = ‘\033[90m’
# Default = ‘\033[99m’

randomTweet = positiveTweets[random.randint(0, 5000)]

# for testing purpose below text is used
# randomTweet = 'testing #python #nltk and give us feedback @https://feedback.com'

print('\n', randomTweet)

# removing hyperlinks
randomTweet = re.sub(r'https?:\/\/.*[\r\n]*', '', randomTweet)
randomTweet = re.sub(r'#', '', randomTweet)
print(randomTweet)

tokenizer = TweetTokenizer(preserve_case=False)
tokens = tokenizer.tokenize(randomTweet)

print('\nTokenized String')
print(tokens)

stopWords = stopwords.words('english')
print('\nStop words: ', stopWords)

punctuations = string.punctuation

cleanTweet = []
for word in tokens:
  if word not in stopWords and word not in punctuations:
    cleanTweet.append(word)

print('\nClean tweet: ', cleanTweet)

stemmer = PorterStemmer()
tweetStems = []

for word in cleanTweet:
  tweetStems.append(stemmer.stem(word))

print('\nStemmed words: ', tweetStems)

lemmatizer = WordNetLemmatizer()
tweetsLem = []

for word in cleanTweet:
  tweetsLem.append(lemmatizer.lemmatize(word))

print('\nLemmatized tweet: ', tweetsLem)

drive.mount('/content/drive')
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mtcars.csv')

d = pd.crosstab(index=[data["mpg"], data["cyl"]], columns="", dropna=False)
# print('\n', d)
print('\n')

data.describe()
data.info()

print('\n')
plt.hist(data["mpg"], bins=3)
plt.show()

print('\n')
plt.scatter(data["wt"], data["mpg"])
plt.show()

```
